```python
import os, shutil
```

```python
AUTO_ML_RUNS_PATH = './runs/final_hyperparam_runs_inside_docker_auto_dl'
AUTO_ML_RESULTS_COPY_PATH = './runs/Cropped_auto_ml_results'

AUTO_ML_RUNS = os.listdir(AUTO_ML_RUNS_PATH)

for counter, run_name in enumerate(AUTO_ML_RUNS, start=1):
    AUTO_ML_RUN_PATH = AUTO_ML_RUNS_PATH + '/' + run_name + '/' + 'train_run'

    # AUTO_ML_RUN_PATH = AUTO_ML_RUN_PATH if os.path.exists(AUTO_ML_RUN_PATH) else AUTO_ML_RUN_PATH.split('train_run')[0] + ''

    AUTO_ML_COPY_PATH = AUTO_ML_RESULTS_COPY_PATH + '/' + run_name
    AUTO_ML_CSV_SRC_PATH = AUTO_ML_RUN_PATH + '/' + 'results.csv'

    print(f'[{counter}] Lookin at {AUTO_ML_RUN_PATH}')

    if os.path.exists(AUTO_ML_CSV_SRC_PATH):
        os.makedirs(AUTO_ML_COPY_PATH, exist_ok=True)
        shutil.copy(AUTO_ML_RUN_PATH + '/' + 'results.csv', AUTO_ML_COPY_PATH + f'/{run_name}.csv')

        print('Completed\n')

    else:
        print(f'[!!] results.csv NOT FOUND IN {AUTO_ML_RUN_PATH}\n')



```
